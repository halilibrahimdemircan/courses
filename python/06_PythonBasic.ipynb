```python
text = "{}, {}, {}".format(3.1213, 5.2332, 7.231243)
print(text)
```

```python
text = "{:.2f}, {:.3f}, {:.4f}".format(3.1213, 5.2332, 7.231243)
print(text)
# nokta formatlama ile verieri düzenleyebiliriz
```

```python
text = "{:.2s}, {:.3s}, {:.4s}".format("3.hgkjg1213", "5bmnö.2332", "7.pioğo231243")
print(text) # string formutanında karakter sayısını belirleriz
```

```python
text = "{:.2}, {:.3}, {:.4}".format("3.hgkjg1213", "5bmnö.2332", "7.pioğo231243")
# s yi kaldırırsak otomatik olarak strin galgılar
```

```python
text = "{:>10}".format("test")
text # 10 karakterlik bir alanda sağa yaslama yaptık
```

```python
text = "bla bla {:<15} formatlama cümleciği".format("Python")
text #15 karakterlik alanda sola yasladık
```

```python
text = "{:^20}".format("merhaba")
text # 20 karakterlik alanda ortaladık
```

```python
text = "{:10.5}".format("dalgakıran")
text # 10 karakterlik alanda ilk beş karakteri aldık
```

```python
"{:>15.5}".format("sdgsgsfgsdfgdfgdfg")
```

```python
"{:+d}".format(110) #girilen integer değerinin başına + veya - ekleme, +d deki d "digit"
```
